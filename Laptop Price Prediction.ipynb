Weight Categories vs Price

Create Weight Categories

```python
df['Weight_category'] = pd.cut(
    df['Weight'],
    bins=[0, 1.5, 2.5, 5],
    labels=['Light (<1.5kg)', 'Medium (1.5–2.5kg)', 'Heavy (>2.5kg)']
)
```

Average Price by Weight Category

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Weight_category',
    y='Price_euros',
    hue='Weight_category',
    data=df,
    estimator=np.mean,
    errorbar=None,
    palette='viridis',
    legend=False
)

plt.title('Average Laptop Price by Weight Category')
plt.xlabel('Weight Category')
plt.ylabel('Average Price (Euros)')
plt.show()
```

Lightweight laptops tend to be more expensive due to premium materials and portability, while heavy laptops are often gaming or workstation devices.
